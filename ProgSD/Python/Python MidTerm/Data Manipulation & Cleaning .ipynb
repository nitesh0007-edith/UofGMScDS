```python
"""
COMPLETE DATA CLEANING AND MANIPULATION TUTORIAL FOR BEGINNERS
===============================================================
This tutorial covers everything you need to know about data cleaning 
and manipulation using pandas in Python.
"""

import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# ============================================================================
# PART 1: DATA LOADING
# ============================================================================

print("=" * 80)
print("PART 1: DATA LOADING")
print("=" * 80)

# Creating sample messy dataset
data = {
    'Name': ['John Doe', 'jane smith', 'BOB JOHNSON', None, 'alice wong', 'Charlie Brown', 'david lee'],
    'Age': [25, 30, None, 45, 28, '35', 'invalid'],
    'Email': ['john@email.com', 'JANE@EMAIL.COM', 'bob@email', None, 'alice@email.com', 'charlie@email.com', ''],
    'Salary': [50000, 60000, 75000, None, 55000, 65000, 70000],
    'Department': ['Sales', 'sales', 'Marketing', 'HR', 'Sales', 'marketing', 'HR'],
    'Join_Date': ['2020-01-15', '2019-06-20', '2021-03-10', None, '2020/08/25', '2018-12-01', '2022-02-14'],
    'Performance': [85, 92, 78, None, 88, 95, None]
}

df = pd.DataFrame(data)
print("\n1. Original Dataset:")
print(df)
print(f"\nDataset shape: {df.shape}")

# ============================================================================
# PART 2: INITIAL DATA EXPLORATION
# ============================================================================

print("\n" + "=" * 80)
print("PART 2: INITIAL DATA EXPLORATION")
print("=" * 80)

print("\n2. Data Info:")
print(df.info())

print("\n3. Data Types:")
print(df.dtypes)

print("\n4. First few rows:")
print(df.head(3))

print("\n5. Last few rows:")
print(df.tail(3))

print("\n6. Statistical Summary:")
print(df.describe())

print("\n7. Check for missing values:")
print(df.isnull().sum())

print("\n8. Check for duplicate rows:")
print(f"Number of duplicate rows: {df.duplicated().sum()}")

# ============================================================================
# PART 3: HANDLING MISSING VALUES
# ============================================================================

print("\n" + "=" * 80)
print("PART 3: HANDLING MISSING VALUES")
print("=" * 80)

df_clean = df.copy()

print("\n9. Missing values percentage:")
missing_pct = (df_clean.isnull().sum() / len(df_clean)) * 100
print(missing_pct[missing_pct > 0])

# Method 1: Drop rows with any missing values
df_dropped = df_clean.dropna()
print(f"\n10. After dropping rows with missing values: {df_dropped.shape}")

# Method 2: Drop rows where all values are missing
df_clean = df_clean.dropna(how='all')

# Method 3: Fill missing values with specific value
df_clean['Name'] = df_clean['Name'].fillna('Unknown')

# Method 4: Forward fill (use previous value)
df_clean['Performance'] = df_clean['Performance'].fillna(method='ffill')

# Method 5: Fill with mean/median/mode
df_clean['Salary'] = df_clean['Salary'].fillna(df_clean['Salary'].median())

print("\n11. After handling missing values:")
print(df_clean)

# ============================================================================
# PART 4: DATA TYPE CONVERSION
# ============================================================================

print("\n" + "=" * 80)
print("PART 4: DATA TYPE CONVERSION")
print("=" * 80)

# Convert Age to numeric (errors='coerce' converts invalid values to NaN)
df_clean['Age'] = pd.to_numeric(df_clean['Age'], errors='coerce')

# Fill NaN ages with median
df_clean['Age'] = df_clean['Age'].fillna(df_clean['Age'].median())

# Convert Age to integer
df_clean['Age'] = df_clean['Age'].astype(int)

print("\n12. After converting Age to numeric:")
print(df_clean['Age'])
print(f"Age data type: {df_clean['Age'].dtype}")

# ============================================================================
# PART 5: STRING CLEANING AND MANIPULATION
# ============================================================================

print("\n" + "=" * 80)
print("PART 5: STRING CLEANING AND MANIPULATION")
print("=" * 80)

# Convert all names to title case
df_clean['Name'] = df_clean['Name'].str.title()

# Convert emails to lowercase
df_clean['Email'] = df_clean['Email'].str.lower()

# Strip whitespace
df_clean['Name'] = df_clean['Name'].str.strip()

# Standardize department names
df_clean['Department'] = df_clean['Department'].str.capitalize()

print("\n13. After string cleaning:")
print(df_clean[['Name', 'Email', 'Department']])

# Replace empty strings with NaN
df_clean['Email'] = df_clean['Email'].replace('', np.nan)

# Check if email is valid (contains @ and .)
df_clean['Valid_Email'] = df_clean['Email'].str.contains('@', na=False) & df_clean['Email'].str.contains('.', na=False)

print("\n14. Email validation:")
print(df_clean[['Email', 'Valid_Email']])

# ============================================================================
# PART 6: DATE HANDLING
# ============================================================================

print("\n" + "=" * 80)
print("PART 6: DATE HANDLING")
print("=" * 80)

# Standardize date format (replace / with -)
df_clean['Join_Date'] = df_clean['Join_Date'].str.replace('/', '-')

# Convert to datetime
df_clean['Join_Date'] = pd.to_datetime(df_clean['Join_Date'], errors='coerce')

print("\n15. After date conversion:")
print(df_clean['Join_Date'])

# Extract date components
df_clean['Join_Year'] = df_clean['Join_Date'].dt.year
df_clean['Join_Month'] = df_clean['Join_Date'].dt.month
df_clean['Join_Day'] = df_clean['Join_Date'].dt.day

# Calculate tenure (days since joining)
df_clean['Tenure_Days'] = (pd.Timestamp.now() - df_clean['Join_Date']).dt.days

print("\n16. Date components:")
print(df_clean[['Join_Date', 'Join_Year', 'Join_Month', 'Tenure_Days']])

# ============================================================================
# PART 7: HANDLING DUPLICATES
# ============================================================================

print("\n" + "=" * 80)
print("PART 7: HANDLING DUPLICATES")
print("=" * 80)

# Add a duplicate row for demonstration
df_with_dup = pd.concat([df_clean, df_clean.iloc[[0]]], ignore_index=True)

print(f"\n17. Duplicates before removal: {df_with_dup.duplicated().sum()}")

# Remove duplicates
df_clean = df_with_dup.drop_duplicates()

print(f"18. Duplicates after removal: {df_clean.duplicated().sum()}")

# Remove duplicates based on specific columns
df_clean = df_clean.drop_duplicates(subset=['Name', 'Email'], keep='first')

# ============================================================================
# PART 8: FILTERING AND SELECTING DATA
# ============================================================================

print("\n" + "=" * 80)
print("PART 8: FILTERING AND SELECTING DATA")
print("=" * 80)

# Select specific columns
print("\n19. Select specific columns:")
print(df_clean[['Name', 'Age', 'Salary']])

# Filter rows based on condition
print("\n20. Filter: Age > 30")
print(df_clean[df_clean['Age'] > 30])

# Multiple conditions
print("\n21. Filter: Age > 25 AND Department = Sales")
filtered = df_clean[(df_clean['Age'] > 25) & (df_clean['Department'] == 'Sales')]
print(filtered)

# Using isin() for multiple values
print("\n22. Filter: Department in ['Sales', 'Hr']")
print(df_clean[df_clean['Department'].isin(['Sales', 'Hr'])])

# Using query() method
print("\n23. Using query method:")
print(df_clean.query('Age > 30 and Salary > 60000'))

# ============================================================================
# PART 9: SORTING DATA
# ============================================================================

print("\n" + "=" * 80)
print("PART 9: SORTING DATA")
print("=" * 80)

# Sort by single column
print("\n24. Sort by Age:")
print(df_clean.sort_values('Age')[['Name', 'Age', 'Salary']])

# Sort by multiple columns
print("\n25. Sort by Department and Salary:")
print(df_clean.sort_values(['Department', 'Salary'], ascending=[True, False])[['Name', 'Department', 'Salary']])

# ============================================================================
# PART 10: CREATING NEW COLUMNS
# ============================================================================

print("\n" + "=" * 80)
print("PART 10: CREATING NEW COLUMNS")
print("=" * 80)

# Simple calculation
df_clean['Annual_Bonus'] = df_clean['Salary'] * 0.10

# Conditional column creation
df_clean['Salary_Grade'] = df_clean['Salary'].apply(
    lambda x: 'High' if x > 65000 else ('Medium' if x > 55000 else 'Low')
)

# Using np.where() for conditions
df_clean['Senior'] = np.where(df_clean['Age'] > 35, 'Yes', 'No')

print("\n26. New columns created:")
print(df_clean[['Name', 'Salary', 'Annual_Bonus', 'Salary_Grade', 'Senior']])

# ============================================================================
# PART 11: GROUPING AND AGGREGATION
# ============================================================================

print("\n" + "=" * 80)
print("PART 11: GROUPING AND AGGREGATION")
print("=" * 80)

# Group by single column
print("\n27. Average salary by department:")
print(df_clean.groupby('Department')['Salary'].mean())

# Multiple aggregations
print("\n28. Multiple aggregations:")
agg_result = df_clean.groupby('Department').agg({
    'Salary': ['mean', 'min', 'max'],
    'Age': 'mean',
    'Name': 'count'
})
print(agg_result)

# Custom aggregation
print("\n29. Custom aggregation:")
print(df_clean.groupby('Department').agg({
    'Salary': lambda x: x.max() - x.min(),
    'Age': 'mean'
}))

# ============================================================================
# PART 12: MERGING AND JOINING
# ============================================================================

print("\n" + "=" * 80)
print("PART 12: MERGING AND JOINING")
print("=" * 80)

# Create sample dataframes for merging
dept_info = pd.DataFrame({
    'Department': ['Sales', 'Hr', 'Marketing'],
    'Manager': ['Alice', 'Bob', 'Charlie'],
    'Budget': [100000, 80000, 90000]
})

print("\n30. Original dataframes:")
print("Employees:")
print(df_clean[['Name', 'Department', 'Salary']].head())
print("\nDepartment Info:")
print(dept_info)

# Inner join
merged = pd.merge(df_clean, dept_info, on='Department', how='inner')
print("\n31. After inner join:")
print(merged[['Name', 'Department', 'Manager', 'Budget']].head())

# Left join
merged_left = pd.merge(df_clean, dept_info, on='Department', how='left')
print("\n32. After left join:")
print(merged_left[['Name', 'Department', 'Manager']].head())

# ============================================================================
# PART 13: RESHAPING DATA
# ============================================================================

print("\n" + "=" * 80)
print("PART 13: RESHAPING DATA")
print("=" * 80)

# Pivot table
print("\n33. Pivot table - Average salary by department:")
pivot = df_clean.pivot_table(
    values='Salary',
    index='Department',
    aggfunc='mean'
)
print(pivot)

# ============================================================================
# PART 14: HANDLING OUTLIERS
# ============================================================================

print("\n" + "=" * 80)
print("PART 14: HANDLING OUTLIERS")
print("=" * 80)

# Detect outliers using IQR method
Q1 = df_clean['Salary'].quantile(0.25)
Q3 = df_clean['Salary'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"\n34. Outlier bounds for Salary:")
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")

# Identify outliers
outliers = df_clean[(df_clean['Salary'] < lower_bound) | (df_clean['Salary'] > upper_bound)]
print(f"\n35. Number of outliers: {len(outliers)}")

# Cap outliers
df_clean['Salary_Capped'] = df_clean['Salary'].clip(lower=lower_bound, upper=upper_bound)

# ============================================================================
# PART 15: RENAMING COLUMNS
# ============================================================================

print("\n" + "=" * 80)
print("PART 15: RENAMING COLUMNS")
print("=" * 80)

# Rename specific columns
df_renamed = df_clean.rename(columns={
    'Name': 'Employee_Name',
    'Age': 'Employee_Age'
})

print("\n36. After renaming columns:")
print(df_renamed.columns.tolist())

# Rename all columns (e.g., lowercase)
df_clean.columns = df_clean.columns.str.lower()
print("\n37. All columns in lowercase:")
print(df_clean.columns.tolist())

# Reset column names back
df_clean.columns = df_clean.columns.str.title()

# ============================================================================
# PART 16: RESETTING AND SETTING INDEX
# ============================================================================

print("\n" + "=" * 80)
print("PART 16: RESETTING AND SETTING INDEX")
print("=" * 80)

# Set Name as index
df_indexed = df_clean.set_index('Name')
print("\n38. After setting Name as index:")
print(df_indexed.head())

# Reset index
df_reset = df_indexed.reset_index()
print("\n39. After resetting index:")
print(df_reset.head())

# ============================================================================
# PART 17: APPLYING FUNCTIONS
# ============================================================================

print("\n" + "=" * 80)
print("PART 17: APPLYING FUNCTIONS")
print("=" * 80)

# Apply function to single column
def categorize_age(age):
    if age < 30:
        return 'Young'
    elif age < 40:
        return 'Mid-Career'
    else:
        return 'Senior'

df_clean['Age_Category'] = df_clean['Age'].apply(categorize_age)

print("\n40. Age categories:")
print(df_clean[['Name', 'Age', 'Age_Category']])

# Apply function to multiple columns
def calculate_tax(row):
    return row['Salary'] * 0.2 if row['Salary'] > 60000 else row['Salary'] * 0.1

df_clean['Tax'] = df_clean.apply(calculate_tax, axis=1)

print("\n41. Tax calculation:")
print(df_clean[['Name', 'Salary', 'Tax']])

# ============================================================================
# PART 18: HANDLING CATEGORICAL DATA
# ============================================================================

print("\n" + "=" * 80)
print("PART 18: HANDLING CATEGORICAL DATA")
print("=" * 80)

# Convert to category type
df_clean['Department'] = df_clean['Department'].astype('category')

# Get category codes
df_clean['Dept_Code'] = df_clean['Department'].cat.codes

print("\n42. Category codes:")
print(df_clean[['Department', 'Dept_Code']].drop_duplicates())

# One-hot encoding
dept_dummies = pd.get_dummies(df_clean['Department'], prefix='Dept')
df_encoded = pd.concat([df_clean, dept_dummies], axis=1)

print("\n43. One-hot encoded departments:")
print(df_encoded[['Name', 'Dept_Hr', 'Dept_Marketing', 'Dept_Sales']].head())

# ============================================================================
# PART 19: EXPORTING CLEANED DATA
# ============================================================================

print("\n" + "=" * 80)
print("PART 19: EXPORTING CLEANED DATA")
print("=" * 80)

# Save to CSV
# df_clean.to_csv('cleaned_data.csv', index=False)
print("\n44. Data can be exported using:")
print("df_clean.to_csv('cleaned_data.csv', index=False)")
print("df_clean.to_excel('cleaned_data.xlsx', index=False)")
print("df_clean.to_json('cleaned_data.json')")

# ============================================================================
# PART 20: FINAL SUMMARY
# ============================================================================

print("\n" + "=" * 80)
print("PART 20: FINAL CLEANED DATASET")
print("=" * 80)

print("\n45. Final cleaned dataset:")
print(df_clean[['Name', 'Age', 'Department', 'Salary', 'Join_Date', 'Performance']])

print("\n46. Final dataset info:")
print(df_clean.info())

print("\n" + "=" * 80)
print("TUTORIAL COMPLETE!")
print("=" * 80)
print("\nKey takeaways:")
print("1. Always explore your data first (info, describe, head, tail)")
print("2. Handle missing values appropriately for your use case")
print("3. Convert data types as needed")
print("4. Clean string data (strip, lowercase, standardize)")
print("5. Handle dates properly using pd.to_datetime()")
print("6. Remove duplicates when necessary")
print("7. Filter and select data efficiently")
print("8. Create new features from existing data")
print("9. Use groupby for aggregations")
print("10. Export cleaned data for further analysis")
```
