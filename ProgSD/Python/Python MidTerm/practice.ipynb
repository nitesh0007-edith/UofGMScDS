```python
import matplotlib.pyplot as plt

# Data: Your tech stack experience
labels = ['Azure Data Factory', 'Databricks (PySpark)', 'Snowflake', 'GenAI']
sizes = [30, 40, 20, 10]  # Hypothetical percentages

# Explode the 2nd slice (PySpark) by 10% of the radius
explode = (0, 0.1, 0, 0) # The value determines how far the slice is pulled out

fig, ax = plt.subplots()
ax.pie(sizes,
       explode=explode,
       labels=labels,
       autopct='%1.1f%%', # Adds percentage labels
       shadow=True,
       startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures the pie is a circle.

plt.title("Technology Experience Distribution")
plt.show()
```
